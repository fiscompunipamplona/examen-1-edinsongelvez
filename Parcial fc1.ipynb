```python
#integrantes DIEGO PALACION-EDINSON GELVEZ 1094277246


from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, cos

pix = 100
size = 100. # cm 
step = size / pix # cm/piXel
lda = 1.0 # cm
k = 2.*pi / lda
zeta0 = 1.0 # cm
X0 = float(input("Escriba el punto del agujero 1:")) # cm
y1 = float(input("Escriba el punto del agujero 2:"))

X = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    X[i] = step*i #COnstruyendo el vector X
    y[i] = step*i #COnstruyendo el vector y
    
onda = open("outdata.dat", "w")

for yy in range( len(X) ):
    for XX in range( len(y) ):
        R1 = sqrt( (X0-X[XX])*(X0-X[XX]) + (y1-y[yy])*(y1-y[yy]) )
        R2 = sqrt( ((X0+20)-X[XX])*((X0+20)-X[XX]) + (y1-y[yy])*(y1-y[yy]) ) 
        zeta1=zeta0*cos(R1*k)
        zeta2=zeta0*cos(R2*k)
        onda.write("%0.1f " % (zeta1))
        onda.write("%0.1f " % (zeta2))
        onda.write("%0.1f " % (zeta1+zeta2))
        
    onda.write("\n")
    
onda.close()

data = loadtxt("outdata.dat", float)
imshow(data,origin="lower",extent=[0,100,0,100])
xlabel("posicin de la onda en X cm")
ylabel("posicin de la onda en y cm")
imshow(data)
show()
```
